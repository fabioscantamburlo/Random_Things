#### HOW TO PLOT PRECISION RECALL F1 SCORE IN A FANCY WAY IN MATPLOTLIB DIFFERENCIATING AMOUNG DIFFERENT CLASSIFIER SYSTEMS IN ONLY ONE GRAPH

This code has the aim of plotting in a "fancy" way the major metrics for evaluating a classification system.


```python
import numpy as np
import matplotlib.pyplot as plt

Rocket_var=[[0.97336066, 0.98339844] ,[0.98241986, 0.97483059],[0.97786927, 0.97909577]
,[0.97643443, 0.98730469], [0.98654244, 0.97775629] ,[0.98146241, 0.98250729]
,[0.9682377  , 0.97949219],[0.97826087, 0.97001934], [0.97322348, 0.97473275]
, [0.98463115, 0.97851562], [0.97761953 ,0.98525074], [0.98111281, 0.98187163]]


Class_1_SVM=[i[0] for i in (Rocket_var[:3])]
Class_2_SVM=[i[1] for i in(Rocket_var[:3])]
Class_1_log=[i[0] for i in (Rocket_var[3:6])]
Class_2_log=[i[1] for i in(Rocket_var[3:6])]
Class_1_Dtree=[i[0] for i in (Rocket_var[6:9])]
Class_2_Dtree=[i[1] for i in(Rocket_var[6:9])]
Class_1_Rfor=[i[0] for i in (Rocket_var[9:12])]
Class_2_Rfor=[i[1] for i in(Rocket_var[9:12])]
```

The vector Rocket_var contains some metrics coming from an evaluation.


The content is this one:

[ [Precision_classifier1_class1 Precision_classifier1_class2], [val val], [F1_classifier1_class1 F1_classifier1_class2], [Precision_classifier2_class1,..],....]


Every three vector inside Rocket var the classifier is changed, and the metrics are PRECISION, RECALL, F1.

```python
labels=['Precision','Recall','F1']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
fig, axes = plt.subplots(2,2,figsize=(16,9),constrained_layout=True) #We want four graphs because we have 4 classifiers

#constrained_layout=True or overlap between suptitle and titles of the subplots

fig.suptitle('Rocket features extraction', fontsize=25)#Title of the figure

rect1=axes[0,0].bar(x - width/2, Class_1_SVM, width, label='Class_1')
rect2=axes[0,0].bar(x + width/2, Class_2_SVM, width, label='Class_2')
axes[0,0].set_title('SVM',fontsize=18)

rect3=axes[0,1].bar(x - width/2, Class_1_log, width, label='Class_1')
rect4=axes[0,1].bar(x + width/2, Class_2_log, width, label='Class_2')
axes[0,1].set_title('LReg',fontsize=18)

rect5=axes[1,0].bar(x - width/2, Class_1_Dtree, width, label='Class_1')
rect6=axes[1,0].bar(x + width/2, Class_2_Dtree, width, label='Class_2')
axes[1,0].set_title('Dtree',fontsize=18)

rect7=axes[1,1].bar(x - width/2, Class_1_Rfor, width, label='Class_1')
rect8=axes[1,1].bar(x + width/2, Class_2_Rfor, width, label='Class_2')
axes[1,1].set_title('Rfor',fontsize=18)

# Add some text for labels, title and custom x-axis tick labels, etc.
for ax in axes.flat:
    ax.set_ylabel('Scores',fontsize=18)
    ax.set_ylim(0,1.1)
    #ax.set_title('Metrics')
    ax.set_xticks(x)
    ax.set_xticklabels(labels,fontsize=18)
    ax.legend(loc=4)

#Autolabel taken from the plotbar documentation matplotlib
def autolabel(rects, ax):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:0.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',fontsize=15)

autolabel(rect1,axes[0,0])
autolabel(rect2,axes[0,0])
autolabel(rect3,axes[0,1])
autolabel(rect4,axes[0,1])
autolabel(rect5,axes[1,0])
autolabel(rect6,axes[1,0])
autolabel(rect7,axes[1,1])
autolabel(rect8,axes[1,1])

#fig.savefig('Rocket_metrics.png')
plt.show()
```
